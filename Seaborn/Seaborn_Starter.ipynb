```python
penguins = sns.load_dataset('penguins')
sns.jointplot(data=penguins, x='flipper_length_mm', y='bill_length_mm', hue='species')
```

```python
sns.pairplot(data=penguins, hue='species')
```

```python
g = sns.PairGrid(penguins, hue='species', corner=True)
g.map_lower(sns.kdeplot, hue=None, levels=5, color='grey')
g.map_lower(sns.scatterplot, marker='+')
g.map_diag(sns.histplot, element='step', kde=True, linewidth=0)
g.add_legend(frameon=True)
g.legend.set_bbox_to_anchor((.61, .6))
```

```python
sns.relplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='body_mass_g')
```

```python
g = sns.relplot(data=penguins, hue='body_mass_g', x='bill_length_mm', y='bill_depth_mm', palette='crest', marker='x', s=100)
g.set_axis_labels('Bill Length (mm)', 'Bill Depth (mm)', labelpad=10)
g.legend.set_title('Body Mass (g)')
g.figure.set_size_inches(6.5, 4.5)
g.ax.margins(.15)
g.despine(trim=True)
```

```python
import matplotlib.pyplot as plt
fig, axs = plt.subplots(1,2, figsize=(8,4), gridspec_kw=dict(width_ratios=[4,3]))
sns.scatterplot(data=penguins, x='flipper_length_mm', y='bill_length_mm', ax = axs[0], hue='species')
sns.histplot(data=penguins, x='species', hue='species', ax=axs[1], legend=False, alpha=.8, shrink=.8)
fig.tight_layout()
```
